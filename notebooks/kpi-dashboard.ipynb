```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/company_financials.csv")

# Calculate KPIs
df["Revenue_Growth_%"] = df["Revenue"].pct_change() * 100
df["Profit_Margin_%"] = (df["Net_Income"] / df["Revenue"]) * 100
df["Expense_Ratio_%"] = (df["Expenses"] / df["Revenue"]) * 100

print(df)
```

```python
plt.figure(figsize=(10,6))

plt.plot(df["Year"], df["Profit_Margin_%"], marker="o", label="Profit Margin")
plt.plot(df["Year"], df["Expense_Ratio_%"], marker="o", label="Expense Ratio")

plt.title("Financial KPI Trends")
plt.xlabel("Year")
plt.ylabel("Percentage")
plt.legend()
plt.grid(True)

plt.savefig("kpi_chart.png", bbox_inches="tight")
plt.show()
```
